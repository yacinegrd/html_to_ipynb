#### Simulation de choix aléatoires[¶](#Simulation-de-choix-al%C3%A9atoires)



```python
malist=["pomme","orange","melon","fraise","banane", "poire"]

w1=rd.choice(malist)  #choisir alétoirement 1 élément de malist
print('w1=',w1)

w2=rd.choice(malist, 4)  #choisir alétoirement 4 éléments de malist (avec répétition permise)
print('w2=',w2)

w3=rd.choice(malist, 4, replace=False)  #choisir alétoirement 4 éléments de malist (sans répétition)
print('w3=',w3)
```

### Exemple1: Simulation des commandes d'un restaurant[¶](#Exemple1:-Simulation-des-commandes-d'un-restaurant)

D'après un sondage dans un restaurant sur 1000 commandes

500 clients ont commandé des sandwiches  
300 clients ont commandé des tacos  
150 clients ont commandé des pizzas  
50 clients ont commandé des glaces  


Essayez de simuler les commandes de 100 clients selon ce sondage



```python
plats=['sandwich','tacos','pizza','glace']   #la liste des plats
prob=[0.5, 0.3, 0.15, 0.05]  #la probabilité de chaque plat (500/1000, 300/1000, 150/1000, 50/1000)

#simuler la commande de 100 clients selon les probabilités
n=100
commandes=rd.choice(plats, n, p=prob)  
print('commandes=',commandes)
```

#### Voir l'histogramme de la simulation[¶](#Voir-l'histogramme-de-la-simulation)



```python
plt.hist(commandes)  #dessiner l'histogramme des commandes
plt.title('Nombre de commandes pour chaque plat sur 100 clients simulés')
plt.show()
```

#### Calcul des probabilités de chaque plat à partir de lasimulation[¶](#Calcul-des-probabilit%C3%A9s-de-chaque-plat-%C3%A0-partir-de-la-simulation)



```python
#Calculer le nombre d'occurence de chaque plat sur les 100 commandes simulées
List_plats, occurences = np.unique(commandes, return_counts=True)
print('plats=', List_plats)
print('occurences=', occurences)

for i, plat in enumerate(List_plats):
    print("Le plat %s a été commandé %d fois/%d sa probabilité=%.4f"%(plat, occurences[i], n, occurences[i]/n))
```

On remarque que les probabilités estimées de la simulation sontproches des probabilités qu'on a spécifié au départ, elles peuventêtre encore plus proches si on augmente la taille du tirage (ex:1000 ou 10000 au lieu de 100)

### Exemple 2: Simulation de l'arrivée d'un évènement[¶](#Exemple-2:-Simulation-de-l'arriv%C3%A9e-d'un-%C3%A9v%C3%A8nement)

Dans un commerce ouvert de 9h à 17h on a observé l'arrivée de 96clients, simulez l'arrivée des clients par minute pour une duréede 2 heures.

#### Solution:[¶](#Solution:)

De 9h à 17h = 8 heures   
La moyenne d'arrivée par minute = 96 clients / (8 heures x 60minutes) = 0.2 clients / minute



```python
#Arrivée d'un client si retour=True,  p: probabilité d'arrivée
def arrivee_client(p): 
    return rd.choice([True, False],p=[p, 1-p])

#fonction de simulation, p: probabilité d'arrivée, duree: durée de la simulation
def simulation_clients(p, duree):
    id_client=0  #numéro du client
    for t in range(duree):
        if arrivee_client(p):
            print("Minute %d: arrivée du client%d"%(t, id_client))
            id_client=id_client+1
            
            

############## Appel de la fonction simulation ############## 
p_arrive=0.2   # la probabilité d'arrivée d'un client par minute
duree=2*60     # durée de la simulation: 2heures x 60minutes
simulation_clients(p_arrive, duree)
```

### Exemple 3: Simulation de la durée d'un évènement[¶](#Exemple-3:-Simulation-de-la-dur%C3%A9e-d'un-%C3%A9v%C3%A8nement)

Sur la boite d'une ampoule il est inscrit: durée de vie moyenne =30 jours. Un client achète 10 ampoules, simulez la durée de vie dechaque ampoule.

#### Solution:[¶](#Solution:)

Durée de vie moyenne d'une ampoule = 30 jours => la probabilitéqu'une ampoule grille par jour = 1/30



```python
#Une ampoule grille si retour=True
def ampoule_grille(p): 
    return rd.choice([True, False],p=[p, 1-p])


def simulation_ampoule(p, nb=1):
    id_ampoule=0 #numéro de l'ampoule à simuler
    jour=0       #compteur de jours
    installation=[0]  #liste pour stocker le jour d'installation de chaque ampoule
    while id_ampoule<=nb:
        if ampoule_grille(p):
            print("Jour %d: L'ampoule %d/%d a grillé, durée de vie: %d jours"
                  %(jour, id_ampoule+1, nb, jour-installation[id_ampoule]))
            id_ampoule=id_ampoule+1
            installation.append(jour)

        jour=jour+1


############## Appel de la fonction simulation ############## 
nb_ampoules=10  #nombre d'ampoules à simuler
p_grille=1/30   #la probabilité qu'une ampoule grille par jour
simulation_ampoule(p_grille, nb=nb_ampoules)
```
